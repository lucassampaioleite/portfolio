1. Crie um arquivo notas_estudantes.txt com as seguintes notas dos alunos de uma turma:

jose 10 15 20 30 40
pedro 23 16 19 22
suzana 8 22 17 14 32 17 24 21 2 9 11 17
giselda 12 28 21 45 26 10
joao 14 32 25 16 89

a) Usando o arquivo texto notas_estudantes.txt escreva um programa que imprime o nome dos alunos que têm mais de seis notas. 

b) Usando o arquivo texto notas_estudantes.txt,  escreva um programa que calcula a média das notas de cada estudante e imprime o nome e a média de cada estudante.

c) Usando o arquivo texto notas_estudantes.txt, escreva um programa que calcula a nota mínima e máxima de cada estudante e imprima o nome de cada aluno junto com a sua nota máxima e mínima.


```python
lista_notas = [
    "jose 10 15 20 30 40\n",
    "pedro 23 16 19 22\n",
    "suzana 8 22 17 14 32 17 24 21 2 9 11 17\n",
    "giselda 12 28 21 45 26 10\n",
    "joao 14 32 25 16 89"
]

with open ("notas_estudantes.txt", "w") as arquivo:  #encoding="utf-8"
    arquivo.writelines(lista_notas)

with open ("notas_estudantes.txt", "r") as arquivo:
    
    print("a) Alunos com mais de 6 notas: ")
    for linha in arquivo:
        valores = linha.split()
        if len(valores) > 7:
            print(f"{valores[0].capitalize()}")
    
    print("b) Alunos e suas médias: ")
    arquivo.seek(0) 
    for linha in arquivo:
        valores = linha.split()
        soma = 0
        for valor in range(1, len(valores)):
            soma += float(valores[valor])
        print(f"{valores[0].capitalize()} possui média {soma/(len(valores)-1):.2f}")

    print("c) Alunos e suas médias: ")
    arquivo.seek(0)
    for linha in arquivo:
        valores = linha.split()
        max = int(valores[1])
        min = int(valores[1])
        for valor in range(2, len(valores)):
            if int(valores[valor]) > max: 
                max = int(valores[valor])
            if int(valores[valor]) < min: 
                min = int(valores[valor])
        print(f"{valores[0].capitalize()} possui nota máxima {max} e {min} como nota mínima")
```

2. Escreva um programa que abra (modo a) um arquivo chamado alunos.txt e permita ao usuário digitar nomes de alunos. Cada nome deve ser adicionado em uma nova linha, sem apagar os nomes já existentes.

```python
with open("alunos.txt", "a") as arquivo:
    while True:
        nome = input("Digite o nome do aluno ou pressione 0 para sair: ")
        if nome == "0":
            break
        arquivo.write(f"{nome}\n")

print("Nomes adicionados com sucesso em alunos.txt!")
```

```python
alunos = []
while True:
        nome = input("Digite o nome do aluno ou pressione 0 para sair: ")
        if nome == "0":
            break
        alunos.append(f"{nome}\n")
with open("alunos.txt", "a") as arquivo:
      arquivo.writelines(alunos)
    

print("Nomes adicionados com sucesso em alunos.txt!")
```

3. Crie um programa que grave em um arquivo chamado notas.txt as notas de 5 alunos digitadas pelo usuário, uma em cada linha. Utilize o método writelines() para inserir todas as linhas de uma vez.

```python
notas = []

for i in range(5):
    nota = input(f"Digite a nota do aluno {i+1}: ")
    notas.append(f"Aluno {i+1}: {nota}\n")

with open("notas.txt", "w") as arquivo:
    arquivo.writelines(notas)

print("Notas gravadas com sucesso em notas.txt!")
```
